```python
import pandas as pd;
```

```python
df = pd.read_csv('adult.csv', index_col=False)

```

```python
df.drop(['fnlwgt'], axis=1, inplace=True)
```

```python
df.head()
```

```python
import numpy as np
df.replace('?', np.nan, inplace=True)
```

```python
df['workclass'].fillna(df['workclass'].mode()[0], inplace=True)
```

```python
df.head(50)
```

```python
for col in df.columns:
    print(f"{col}: {df[col].nunique()}")
```

```python
df['sex'].replace(['Male','Female'] , [0,1] , inplace=True)
df['income'].replace(['<=50K','>50K'] , [0,1] , inplace=True)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
workclass_encoder = LabelEncoder()
education_encoder = LabelEncoder()
marital_encoder = LabelEncoder()
occupation_encoder = LabelEncoder()
rel_encoder = LabelEncoder()
race_encoder = LabelEncoder()
native_encoder = LabelEncoder()

df['workclass'] = workclass_encoder.fit_transform(df['workclass'])
df['education'] = education_encoder.fit_transform(df['education'])
df['marital.status'] = marital_encoder.fit_transform(df['marital.status'])
df['occupation'] = occupation_encoder.fit_transform(df['occupation'])
df['relationship'] = rel_encoder.fit_transform(df['relationship'])
df['race'] = race_encoder.fit_transform(df['race'])
df['native.country'] = native_encoder.fit_transform(df['native.country'])


```

```python
x=df.drop(['income'], axis=1)
y=df['income']
```

```python
xtrain, xtest , ytrain , ytest = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB

modelDT = DecisionTreeClassifier()
modelNB = GaussianNB()
```

```python
modelDT.fit(xtrain , ytrain)
modelDT.score(xtest , ytest)
```

```python
modelNB.fit(xtrain , ytrain)
modelNB.score(xtest , ytest)
```
